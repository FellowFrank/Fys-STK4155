## Diagonalization of the full matrix (ex. 2)

```python
for g in [-1,-0.5,0,0.5,1]:
    Matrix = np.array([[2 - g, -g / 2, -g / 2, -g / 2, -g / 2,0],[-g/2,4-g,-g/2,-g/2,0,-g/2],[-g/2,-g/2,6-g,0,-g/2,-g/2],[-g/2,-g/2,0,6-g,-g/2,-g/2],[-g/2,0,-g/2,-g/2,8-g,-g/2],[0,-g/2,-g/2,-g/2,-g/2,10-g]])
    eigenvalues, eigenvectors = np.linalg.eigh(Matrix)
    idx = eigenvalues.argsort()
    sorted_eigenvalues = eigenvalues[idx]
    
    print(f"\n--- g = {g} ---")
    print(f"Ground State Energy (E1): {sorted_eigenvalues[0]:.4f}")
    print(f"Sorted Eigenvalues (E1 to E6): {sorted_eigenvalues}")
```

### Calculating the Energies with the correction 
$E \sim E + E^{(2)} + E^{(3)}$

```python
for g in [-1,-0.5,0,0.5,1]:
    E_0 = 2 - g 
    E_2 = - (g**2 / 4)*(2/(4 + g) + 1/(6 + g) + 1/(2 + g))
    E_3_1 = - (g**3 / 8)*( (1/(g+4) + 1/(g+6))**2 +(1/(g+2) + 1/(g+4))**2)
    E_3_2 = - (g**3 / 8)*( 1/(g+6)**2 + 1/(g+2)**2 + 2/(g+4)**2 + 2/((g+4)*(g+6)) + 2/((g+2)*(g+4)))
    print(f"\n--- g = {g} ---")
    print(f"Perturbation Theory E0: {E_0:.4f}")
    print(f"Perturbation Theory E2: {E_2:.4f}")
    print(f"Perturbation Theory E3: {E_3_1 + E_3_2:.4f}")
    print(f"Perturbation Theory E3 analysis: {E_3_1:.4f}")  
    print(f"Perturbation Theory E3 cross terms: {E_3_2:.4f}")
    print(f"Perturbation Theory E: {E_0 + E_2 + E_3_1 + E_3_2:.4f}")
```
